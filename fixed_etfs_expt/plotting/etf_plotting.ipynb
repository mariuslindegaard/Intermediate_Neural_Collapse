```python
import numpy as np
import matplotlib.pyplot as plt
import pickle
import seaborn as sns
import pandas as pd
```

```python
with open('mnist_etfs_accuracy.pkl','rb') as f:
    acc = pickle.load(f)

df = pd.DataFrame(acc)
df = df.rename(columns={"train_acc": "Train", "test_acc": "Test"})
```

```python
sns.set_style('darkgrid')
# sns.color_palette("rocket_r")
fig, ax = plt.subplots(1, 1, figsize=(4, 4))
axes = (ax,)
plt.sca(axes[0])
sns.lineplot(data=df, palette='rocket', marker='s')
plt.xlabel('Layers Trained')
plt.ylabel('Accuracy')
plt.gca().get_legend()
ax.set_xticks(np.arange(2,11))

plt.title('MNIST MLP with ETFs')
plt.tight_layout()
plt.savefig('mnist_mlp.pdf')
```
